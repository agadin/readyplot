# GROUP DATA


```python
# Import column titles
ColumnTitles = DATA.columns.tolist()
print(ColumnTitles)
```

```python
# Extract feature names
Features = ColumnTitles[:5]
print(Features)
```

```python
# Extract variable names
Vars = ColumnTitles[5:]
print(Vars)
```

```python
# Set up a units list for future y labels
Units = ['cm^2','m','Pa','Pa','','Pa','','Pa','','Pa','','Pa','','Pa','Pa','N','kN','dm','MN','hm','MN','mm','kN','mm','N','','N','N','kJ']
```

# VISUALIZE DATA

```python
whisk_plotter = readyplot.boxwhisker_plotter.BoxWhiskerPlotter('box_strip',DATA,'Group','Var1','Feature3',colors=['c','m','g'],
                                dodge=True,folder_name = 'OUTPUT_FIGURES',low_cap0=True,
                                handles_in_legend = 3,fig_width = 10,box_width = 0.9,
                                custom_y_label ='Var1' + ' (' + Units[Vars.index('Var1')]+')')  
whisk_plotter
whisk_plotter.large_loop()
```

```python
scatter_plotter = readyplot.scatter_plotter.ScatterPlotter('scatter_R2',DATA,'Var2','Var1','Feature3',colors=['c','m','g'],
                                dodge=True,folder_name = 'OUTPUT_FIGURES',low_cap0=True,
                                handles_in_legend = 3,fig_width = 10,box_width = 0.9,
                                custom_y_label ='Var1' + ' (' + Units[Vars.index('Var1')]+')')  
scatter_plotter
scatter_plotter.large_loop()
```
